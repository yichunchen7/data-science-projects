## III Create Schema

```python
# DEFINE / CREATE ALL VERTICES:

conn.gsql('''
    USE GLOBAL
     CREATE VERTEX Patients(PRIMARY_ID id STRING, first_name STRING, last_name STRING, address STRING, DOB STRING, email STRING, gender STRING, phone STRING) WITH primary_id_as_attribute = "true"
     CREATE VERTEX Symptoms(PRIMARY_ID id STRING, symptom STRING) WITH primary_id_as_attribute = "true"
     CREATE VERTEX Diseases(PRIMARY_ID id STRING, disease STRING) WITH primary_id_as_attribute = "true"
     CREATE VERTEX Specialists(PRIMARY_ID id STRING, specialist STRING) WITH primary_id_as_attribute = "true"
''')
```

```python
# DEFINE / CREATE ALL EDGES:
    
conn.gsql('''
   USE GLOBAL
    CREATE DIRECTED EDGE has_symptoms(FROM Patients, TO Symptoms) WITH REVERSE_EDGE="reverse_has_symptoms"
    CREATE DIRECTED EDGE shows_on_disease(FROM Symptoms, TO Diseases) WITH REVERSE_EDGE="reverse_shows_on_disease"
    CREATE DIRECTED EDGE refer_to_specialist(FROM Diseases, TO Specialists) WITH REVERSE_EDGE="reverse_refer_to_specialist"
 ''')
```

## IV Create loading jobs

```python
# create loading job for all vertices: ---need to standardize symtoms
conn.gsql('''
 USE GRAPH UmbrellaAI
 CREATE LOADING JOB load_Vertices FOR GRAPH UmbrellaAI {
    DEFINE FILENAME PatientDataSource;
    DEFINE FILENAME SymptomDataSource;
    DEFINE FILENAME DiseaseDataSource;
    DEFINE FILENAME SpecialistDataSource;
    
    LOAD PatientDataSource TO VERTEX Patients VALUES($0,$1,$2,$5,$6,$3,$4,$7) USING SEPARATOR=",", HEADER="true", EOL="\n", QUOTE="double";
    LOAD SymptomDataSource TO VERTEX Symptoms VALUES($0, $1) USING SEPARATOR=",", HEADER="true", EOL="\n", QUOTE="double";
    LOAD DiseaseDataSource TO VERTEX Diseases VALUES($0, $1) USING SEPARATOR=",", HEADER="true", EOL="\n", QUOTE="double";
    LOAD SpecialistDataSource TO VERTEX Specialists VALUES($0, $1) USING SEPARATOR=",", HEADER="true", EOL="\n", QUOTE="double";
 }
 ''')
```

```python
# create loading job for all edges: 
conn.gsql('''
USE GRAPH UmbrellaAI
CREATE LOADING JOB load_Edges FOR GRAPH UmbrellaAI {
    DEFINE FILENAME patient_symptom_edge_DataSource; 
    DEFINE FILENAME symptom_disease_edge_DataSource;
    DEFINE FILENAME refer_to_edge_DataSource;

    LOAD patient_symptom_edge_DataSource TO EDGE has_symptoms VALUES($0, $3) USING SEPARATOR=",", HEADER="true", EOL="\n", QUOTE="double";
    LOAD symptom_disease_edge_DataSource TO EDGE shows_on_disease VALUES($2, $0) USING SEPARATOR=",", HEADER="true", EOL="\n", QUOTE="double";
    LOAD refer_to_edge_DataSource TO EDGE refers_to_specialist VALUES($0, $2) USING SEPARATOR=",", HEADER="true", EOL="\n", QUOTE="double";
     }
 ''')
```

```python
# find the specialists for patients with symptom "sleeplessness" and disease 'bipolar disorder'
conn.runInterpretedQuery('''
INTERPRET QUERY which_specialist() FOR GRAPH UmbrellaAI {
    start = {Symptoms.*};
    result_specialist = SELECT sp FROM start:s - (shows_on_disease>:e1) - Diseases:d - (refers_to_specialist>:e2) - Specialists:sp
    WHERE s.symptom == "sleeplessness" AND d.disease=="bipolar disorder";
PRINT result_specialist;
}
 ''')
```

```python
# find the diseases that have symptom "sleeplessness"
conn.runInterpretedQuery('''
INTERPRET QUERY which_disease() FOR GRAPH UmbrellaAI {
    start = {Symptoms.*};
    result_disease = SELECT  d FROM start:s - (shows_on_disease>:e) - Diseases:d 
    where s.symptom=="sleeplessness";
PRINT result_disease;
}
 ''')
```

```python
# find the symptoms for disease "hypertensive disease"
conn.runInterpretedQuery("""
INTERPRET QUERY disease_has_symptoms() FOR GRAPH UmbrellaAI {
    start = {Diseases.*};
    symptom_result = SELECT sy FROM start:s - () - Symptoms:sy
    WHERE s.disease == "hypertensive disease";
PRINT symptom_result;
  }
 """)
```

```python
# find all patients who have a particular symptom: for example, "fever"
conn.runInterpretedQuery('''
INTERPRET QUERY patients_with_symptom() FOR GRAPH UmbrellaAI {
    start = {Symptoms.*};
    result_patients_with_symptom = SELECT p FROM start:s - (<has_symptoms:e) - Patients:p
    WHERE s.symptom == "fever";
PRINT result_patients_with_symptom;
}
 ''')
```

```python
# find all patients who have symptoms of a particular disease such as "hypertensive disease"
conn.runInterpretedQuery('''
INTERPRET QUERY patients_with_symptom_of_a_disease() FOR GRAPH UmbrellaAI {
    start = {Diseases.*};
    result_patients_with_symptom_of_a_disease = SELECT p FROM start:s - (<shows_on_disease:e1) - Symptoms:sym - (<has_symptoms:e2) - Patients:p
    WHERE s.disease == "hypertensive disease";
PRINT result_patients_with_symptom_of_a_disease;
}
 ''')
```

```python
# find a patient's symptoms and potential diseases: for example, for patient_id=1
conn.runInterpretedQuery('''
INTERPRET QUERY symptoms_and_diseases_of_patient() FOR GRAPH UmbrellaAI {
    start = {Patients.*};
    result_symptoms_of_patient = SELECT sym FROM start:s - (has_symptoms>:e) - Symptoms:sym
    WHERE  s.id== "1";
    result_diseases_of_patient = SELECT d FROM start:s - (has_symptoms>:e1) - Symptoms:sym - (shows_on_disease>:e2) - Diseases:d
    WHERE  s.id== "1";
PRINT result_symptoms_of_patient;
PRINT result_diseases_of_patient;
}
 ''')
```

```python
# find TOP 10 patients who have multiple symptoms:
conn.runInterpretedQuery('''
INTERPRET QUERY patients_with_multiple_symptoms() FOR GRAPH UmbrellaAI {
    start = {Patients.*};
    SumAccum<int>@symptom_count = 0;
    result_patients_with_multiple_symptoms = SELECT s FROM start:s - (has_symptoms>:e) - Symptoms:sym
                                            ACCUM s.@symptom_count += 1
                                            HAVING s.@symptom_count >= 2
                                            ORDER BY s.@symptom_count DESC
                                            LIMIT 10;
PRINT result_patients_with_multiple_symptoms;
}
 ''')
```

```python
# find total number of symptoms for a disease, such as 'hypertensive diseases':
conn.runInterpretedQuery('''
INTERPRET QUERY symptom_count_of_a_disease() FOR GRAPH UmbrellaAI {
    start = {Diseases.*};
    SumAccum<int>@symptom_count = 0;
    result_symptom_count_of_a_disease = SELECT s FROM start:s - (<shows_on_disease:e) - Symptoms:sym
                                        WHERE s.disease == "hypertensive disease"
                                        ACCUM s.@symptom_count += 1;
PRINT result_symptom_count_of_a_disease;
}
 ''')
```

## standardize symptoms:

```python
# create a function to get symptoms that contain a certain word (such as 'fever')

import re
 
def find_symptoms_with_word(symptom_list, word):
    matched_symptoms = []
    for i in range(len(symptom_list)):
         if re.findall(word, symptom_list[i], re.I):
            matched_symptoms.append(symptom_list[i])
    return matched_symptoms
```
